```python
import time
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

from tangle import Tangle

plt.rc('axes', labelsize=20)
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)
```

Probability density function of tip count

```python
plt.figure(figsize=(15, 8))
t = Tangle(rate=50, tip_selection='urts')

tip_count_1 = []

for i in range(100000):
    t.next_transaction()
    tip_count_1.append([t.time, len(t.tips())])

tip_count_1 = np.array(tip_count_1)
density_1 = stats.kde.gaussian_kde(tip_count_1[:, 1])
```

```python
t = Tangle(rate=50, tip_selection='mcmc', alpha=0.001)

tip_count_2 = []

for i in range(100000):
    t.next_transaction()
    tip_count_2.append([t.time, len(t.tips())])

tip_count_2 = np.array(tip_count_2)
density_2 = stats.kde.gaussian_kde(tip_count_2[:, 1])

x = np.linspace(70, 150, 200)
plt.plot(x, density_1(x), color='red')
plt.plot(x, density_2(x), color='green')
plt.xlabel('Tip count')
plt.ylabel('PDF')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.plot(x, density_1(x), color='red')
plt.plot(x, density_2(x), color='green')
plt.xlabel('Tip count')
plt.ylabel('PDF')
plt.show()
```
