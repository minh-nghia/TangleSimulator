```python
import time
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

from tangle import Tangle

plt.rc('axes', labelsize=20)
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)
```

Tip count growth for higher alpha ($\alpha = 0.1$). There are more unapproved tips due to more deterministic random walk, leaving some tips behind.

```python
plt.figure(figsize=(25, 10))
t = Tangle(rate=50, tip_selection='mcmc', alpha=0.1)

tip_count = []

for i in range(100000):
    t.next_transaction()
    tip_count.append([t.time, len(t.tips())])
        
tip_count = np.array(tip_count)
plt.plot(tip_count[:, 0], tip_count[:, 1])
plt.xlabel('Time')
plt.ylabel('Tip count')
plt.show()
```
